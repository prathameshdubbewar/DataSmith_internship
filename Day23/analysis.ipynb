```python
import pandas as pd
```

```python
df = pd.read_excel('InternData.xlsx')
```

```python
df
```

```python
df['PI (Productivity Index)'] = df['Reservoir Gas Rate (MMSCF/D)'] / (df['SiBHP (psig)'] - df['BHP (psig)'])
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
def calculate_constant_pi(group):

    group['Constant Pi'] = group['PI (Productivity Index)'].iloc[0]

    group['Constant Pi'] = group['Constant Pi'].ffill()
    return group

df = df.groupby(['Well', 'Leg']).apply(calculate_constant_pi)

```

```python
df
```

```python
df['new_const_pi'] = df['Constant Pi'] *1000
```

```python
df['md_diff'] = df.groupby([df['Well'], df['Leg']])['Measured Depth (ft)'].transform(lambda x: x - x.iloc[0])
```

```python
df['Constant PI/ft'] = df['new_const_pi'] / df['md_diff']
```

```python
df
```

##### sol -1

```python
not_zero_kop = df['KOP ft'] != 0

df['Lateral Length'] = df['ROP'] + (df['Measured Depth (ft)'][0] - df['KOP ft'][0]) * not_zero_kop

df.loc[~not_zero_kop, 'Lateral Length'] = 0
```

```python
df
```

```python
df.columns
```

#### sol-2

```python
def calc_lat_len(row):
    if row['KOP ft'] != 0:
        kop_ft_first = df['KOP ft'].iloc[0]
        measured_depth_first = df['Measured Depth (ft)'].iloc[0]
        return row['ROP'] + (measured_depth_first - kop_ft_first)
    else:
        return 0

df['Lateral Length'] = df.apply(calc_lat_len, axis=1)
```

```python
import numpy as np

mask = np.isinf(df['Constant PI/ft'])

df.loc[mask, 'Constant PI/ft'] = np.nan



df['Constant PI/ft'].fillna(method='bfill', inplace=True)
```

```python
value = (df['Constant PI/ft'] * df['Lateral Length']).sum() / df['Lateral Length'].sum()

print("Weighted Constant PI per foot:", value)
```
