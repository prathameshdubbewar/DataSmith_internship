```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('tesla_price.csv')

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
rolling_avg = df['Close'].rolling(window=30).mean()

plt.plot(df.index, df['Close'], label='Original Stock Prices')
plt.plot(rolling_avg.index, rolling_avg, label='30 Day Average')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
